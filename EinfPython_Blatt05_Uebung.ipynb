
### b) Funktionen als Argumente
Betrachten Sie die folgende Funktion `wertetabelle`, die als Argumente eine Funktion und ein iterierbares Objekt von Auswertungsstellen erwartet.  
Erstellen Sie (Python-) Funktionen für die (mathematischen) Funktionen $g\colon x\mapsto x^2$ und $h\colon x \mapsto 5x-x^3$ und verwenden Sie die Funktion `wertetabelle`, um die Wertetabellen dieser Funktionen für $x\in \{-25,-24,\dots,0,\dots,25\}$ auszugeben.

````python
def wertetabelle(f, xe):
    for x in xe:
        print("f({}) = {}".format(x, f(x)))

def g(x):
    return x**2

def h(x):
    return 5*x-x**3

wertetabelle(g, range(-25, 25+1))
wertetabelle(h, range(-25, 25+1))
````

### c) Funktionen als Parameter
Schreiben Sie eine Funktion `my_map(f, iterable)`, deren erster Parameter eine Funktion $f$ ist, für welche die Auswertung $f(x)$ möglich ist, und als zweiten Parameter ein iterierbares Objekt `iterable` besitzt und eine Liste mit den Elementen $\big(f(x)\big)_{x \in \text{iterable}}$ zurückgibt.

Beispielaufruf: `my_map(f, range(3))`  
$\rightarrow$ erwartete Ausgabe: `[f(0), f(1), f(2)]`, also z. B. `[0, 1, 4]` falls $f(x) = x^2$.

````python
def my_map(f, iterable):
    mapped = []
    for x in iterable:
        mapped.append(f(x))
    return mapped

my_map(lambda x: x**2, [0,1,2])
````


#### Anwendung von `my_map`
Verwenden Sie die Funktion `my_map(f, iterable)` indem Sie geeignete Funktionen `f` implementieren, um

(iii) (**Bonus**) zur Menge `FUNCS` eine Liste der Funktionswerte aller Funktionen `f` an **der Stelle** `v=[1,3,3,7]`  

zu erzeugen.

````python
STRS = "Diese Liste von Strings wurde mit der Methode str.split erzeugt, die sehr nützlich ist".split()
NUMS = tuple(range(10, 100, 3))

def prod(iterable):
    res = 1
    for i in iterable:
        res *= i
    return res
FUNCS = [min, sum, max, len, prod]

print(my_map(len, STRS))
print(my_map(lambda num: num**2, NUMS))
print(my_map(lambda fun: fun([1,3,3,7]), FUNCS))
````

### d) Anonyme Funktionen (lambda)
Mithilfe der Anweisung `lambda` können Sie *anonyme Funktionen* definieren. Z.B. erzeugt das folgende Programm eine anonyme Funktion, die die Summe zweier Zahlen zurückgibt:
```python
lambda x, y: x + y
```
Dabei sind `lambda`-Anweisungen selbst *Ausdrücke*, die als rechte Seite einer Zuweisung oder als Argumente bei einem Funktionsaufruf verwendet werden können.

```python
f = lambda x, y: x + y
```
ist äquivalent zu
```python
def f(x, y):
    return x + y
```

*Bemerkung:  
Häufig werden lambda-Ausdrücke benutzt, wenn man nur eine einzeilige Funktion benötigt oder diese nur einmalig verwenden will.*

````python
NUMS = tuple(range(10, 100, 3))

my_map(lambda x: x**2, NUMS)
````

## Aufgabe 2 (Spider Solitaire)
Wir werden in diesem Semester das Spiel Spider Solitaire programmieren. Bitte lesen Sie zunächst die Spielregeln (Spielregeln.pdf). 

Eine Karte besitzt zwei Attribute: Die Farbe (Herz oder Pik) und den Kartenwert (13,...,1). Wir nutzen in dieser Aufgabe zweielementige Tupel, um Karten zu modellieren:

````python
card1 = ("hearts", 3)
card2 = ("hearts", 2)
card3 = ("spades", 2)
card4 = ("hearts", 1)
````

### a) Funktion `fits_to` auf Kartenlevel
**Sinn der Funktion:** Die Funktion `fits_to` soll überprüfen, ob eine Karte an eine weitere Karte angelegt werden kann. 

**Schnittstelle der Funktion:** Beide Karten werden als Positionsargumente übergeben. Zusätzlich soll die Funktion einen optionalen booleschen Parameter `matching_suit` mit dem Default-Wert `True` besitzen. In Standardfall soll die Funktion genau dann `True` liefern, wenn die `value`-Attribute der Karten zusammen passen, d.h. der Wert der ersten Karte um eins niedriger als der von der zweiten Karte ist, und die Farben übereinstimmen. Hat `matching_suit` den Wert `False`, dann müssen die Kartenwerte zusammenpassen und die Farben unterschiedlich sein, um als Rückgabe `True` zu erhalten.

Implementieren Sie die Funktion `fits_to` und testen Sie die Funktionweise mit den bereits vorhandenen Tests. Es sollte vier Mal `True` ausgegeben werden.

````python
# Implementieren Sie hier fits_to
def fits_to(child, parent, matching_suit=True):
    # return false if matching_suit is different to whether the matches suit
    if matching_suit and child[0] != parent[0]: return False
    if not matching_suit and child[0] == parent[0]: return False
    return child[1] + 1 == parent[1]


print(fits_to(card2, card1, matching_suit=False) == False)
print(fits_to(card3, card1, matching_suit=False) == True)
print(fits_to(card2, card1) == True)
print(fits_to(card3, card1) == False)
````

### b) Die Funktion `is_sequence`
**Sinn der Funktion:** Die Funktion soll überprüfen, ob es sich bei einer Liste von Karten um eine Sequenz handelt, also alle Karten von der gleichen Farbe sind und die Kartenwerte eine **nichtleere** absteigende passende Folge bilden.  

**Schnittstelle der Funktion:** Es wird eine Liste von Karten übergeben. Rückgabewert ist der boolsche Wert `True`, falls es sich um eine Sequenz handelt, sonst `False`. 

Implementieren Sie die Funktion `is_sequence` unter Zuhilfenahme der Funktion `fits_to` aus Teilaufgabe a). Testen Sie die Funktionweise mit den bereits vorhandenen Tests. Es sollte drei Mal `True` ausgegeben werden.

````python
listCards1 = [card1, card2]
listCards2 = [card1, card3]
listCards3 = [card1, card2, card4]

# Implementieren Sie hier is_sequence
def is_sequence(cards):
    for i in range(1, len(cards)):
        if not fits_to(cards[i], cards[i-1], True): return False
    return True


print(is_sequence(listCards1) == True)
print(is_sequence(listCards2) == False)
print(is_sequence(listCards3) == True)
````

### c) Die Funktion `fits_to_sequence`
**Sinn der Funktion:** Die Funktion `fits_to_sequence` soll überprüfen, ob eine Sequenz an eine weitere Sequenz angelegt werden kann. 

**Schnittstelle der Funktion:** Beide Sequenzen werden als Positionsargumente übergeben. Zusätzlich soll die Funktion einen optionalen booleschen Parameter `matching_suit` mit dem Default-Wert `True` besitzen, wie es bei der Funktion `fits_to` auf Kartenlevel der Fall ist. Die Bedeutung dieses Default-Parameters ist entsprechend. 

Implementieren Sie die Funktion `fits_to_sequence` und testen Sie die Funktionweise mit den bereits vorhandenen Tests. Es sollte fünf Mal `True` ausgegeben werden.

````python
sequence1 = [("hearts", n) for n in range(8, 4, -1)]
sequence2 = [("hearts", n) for n in range(13, 8, -1)]
sequence3 = [("hearts", n) for n in range(13, 9, -1)]
sequence4 = [("spades", n) for n in range(13, 8, -1)]

# Implementieren Sie die Funktion fits_to_sequence.
def fits_to_sequence(child, parent, matching_suit=True):
    return fits_to(child[0], parent[-1], matching_suit)
    

print(fits_to_sequence(sequence1, sequence2) == True)
print(fits_to_sequence(sequence1, sequence4) == False)
print(fits_to_sequence(sequence1, sequence3) == False)
print(fits_to_sequence(sequence1, sequence3, matching_suit=False) == False)
print(fits_to_sequence(sequence1, sequence4, matching_suit=False) == True)
````
